```python
NOMBRES=['f800.txt','f850.txt','f900.txt','f950.txt','f1000.txt','f1050.txt','f1100.txt','f1150.txt','f1200.txt','f1250.txt','f1300.txt']
frecuencias_max=[]
for string in NOMBRES:
    X=np.loadtxt(string, usecols=0)
    Y=np.loadtxt(string, usecols=1)
    if string=='f900.txt':
        plt.figure(figsize=(10,7))
        plt.plot(X,Y-np.min(Y),label='$f$=900 Hz',c='darkgreen')
        plt.xlabel('$f$ (Hz)', fontsize=15)
        plt.ylabel('A (V)', fontsize=15)
        plt.text(1200,70,'fmax=900.1 Hz', fontsize=15)
        plt.scatter([900.114],[71.6],c='black')
        plt.legend(fontsize=13)
    if string=='f950.txt':
        frecuencias_max.append(X[where(Y==max(Y))][1])
    else:
        frecuencias_max.append(mean(X[where(Y==max(Y))]))
```
