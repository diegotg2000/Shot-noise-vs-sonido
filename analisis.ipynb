```python
NOMBRES=['f800.txt','f850.txt','f900.txt','f950.txt','f1000.txt','f1050.txt','f1100.txt','f1150.txt','f1200.txt','f1250.txt','f1300.txt']
frecuencias_max=[]
for string in NOMBRES:
    X=np.loadtxt(string, usecols=0)
    Y=np.loadtxt(string, usecols=1)
    if string=='f900.txt':
        plt.plot(X,Y-np.min(Y),label=string)
    if string=='f950.txt':
        frecuencias_max.append(X[where(Y==max(Y))][1])
    else:
        frecuencias_max.append(mean(X[where(Y==max(Y))]))
```
